# Práctica 2 – Estadística Descriptiva (Anime Dataset)

En esta práctica vamos a:

- Cargar el dataset limpio `Anime_clean.csv`.
- Obtener estadísticos descriptivos de las variables numéricas.
- Analizar variables categóricas (`Type`, `Release_season`, `Studio`, etc.).
- Agrupar por entidades (Type, año de salida, estudio).
- Obtener estadísticas descriptivas por grupo.
- Proponer entidades y relaciones para un diagrama ER.


```python
import pandas as pd

pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/Anime_clean.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nTipos de datos:")
print(df.dtypes)

print("\nPrimeras filas:")
df.head()
```

```python
numeric_cols = ["Rank", "Episodes", "Rating", "Release_year", "End_year"]

df[numeric_cols].describe().T
```

```python
cat_cols = ["Type", "Release_season", "Studio"]

for col in cat_cols:
    if col in df.columns:
        print(f"\nColumna: {col}")
        print("Valores únicos:", df[col].nunique())
        print(df[col].value_counts().head(10))
```

```python

group_type = df.groupby("Type").agg(
    count=("Name", "count"),
    avg_rating=("Rating", "mean"),
    avg_episodes=("Episodes", "mean"),
    avg_release_year=("Release_year", "mean")
).sort_values("avg_rating", ascending=False)

group_type
```

```python
group_year = df.groupby("Release_year").agg(
    count=("Name", "count"),
    avg_rating=("Rating", "mean"),
    avg_episodes=("Episodes", "mean")
).sort_values("Release_year")

group_year.head(20), group_year.tail(20)
```

```python

if "Studio" in df.columns:
    group_studio = df.groupby("Studio").agg(
        count=("Name", "count"),
        avg_rating=("Rating", "mean"),
        avg_episodes=("Episodes", "mean")
    ).sort_values("count", ascending=False)

    
    group_studio.head(20)

```
