# Práctica 3 – Visualización de Datos (Anime Dataset)

En esta práctica generaremos múltiples visualizaciones a partir del dataset `Anime_clean.csv`, incluyendo:

- Histogramas  
- Boxplots  
- Gráficas de pastel  
- Gráficas de dispersión (scatter plots)  
- Barras agrupadas  
- Visualizaciones generadas con ciclos `for`  

El objetivo es entender las distribuciones, relaciones y patrones del dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

df = pd.read_csv("../data/Anime_clean.csv")

df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Rating"], kde=True, bins=30, color="skyblue")
plt.title("Distribución de Ratings")
plt.xlabel("Rating")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Episodes"], kde=False, bins=40, color="salmon")
plt.title("Distribución de Episodios")
plt.xlabel("Episodios")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Type", y="Rating", data=df, palette="Set2")
plt.title("Rating por Tipo de Anime")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Type", y="Episodes", data=df, palette="Set3")
plt.title("Episodios por Tipo de Anime")
plt.show()
```

```python
type_counts = df["Type"].value_counts()

plt.figure(figsize=(7,7))
plt.pie(type_counts, labels=type_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribución de Tipos de Anime")
plt.show()
```

```python
sns.scatterplot(x="Episodes", y="Rating", data=df, alpha=0.6)
plt.title("Rating vs Episodios")
plt.show()
```

```python
sns.scatterplot(x="Release_year", y="Rating", data=df, alpha=0.6, color="purple")
plt.title("Rating vs Año de Lanzamiento")
plt.show()
```

```python
avg_type = df.groupby("Type")["Rating"].mean().sort_values()

avg_type.plot(kind="bar", color="teal")
plt.title("Rating promedio por Tipo de Anime")
plt.xlabel("Tipo")
plt.ylabel("Rating promedio")
plt.show()
```

```python
numeric_cols = ["Rank", "Episodes", "Rating", "Release_year"]

for col in numeric_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Histograma de {col}")
    plt.show()
```

```python
for col in numeric_cols:
    if col != "Rating":
        plt.figure(figsize=(8,4))
        sns.scatterplot(x=df[col], y=df["Rating"])
        plt.title(f"Rating vs {col}")
        plt.show()
```
